#หาว่าในเเต่ละช่วงเวลา เครื่องไหนสามารถทำยอดขายได้มากที่สุด

```python
lko=t_20171207.merge(o_20171207[['OperatorID','WorkstationID']],how = 'left',on ='OperatorID')
lko1=t_20190213.merge(o_20190213[['OperatorID','WorkstationID']],how = 'left',on ='OperatorID')
lko2=t_20190328.merge(o_20190328[['OperatorID','WorkstationID']],how = 'left',on ='OperatorID')
```

```python
klo=lko[['WorkstationID','ArtNum']]
klo1=lko1[['WorkstationID','ArtNum']]
klo2=lko2[['WorkstationID','ArtNum']]
```

```python
sum_of_ArtNum_20171207 = klo.groupby('WorkstationID').sum().sort_values(by='ArtNum',ascending = False).reset_index()
sum_of_ArtNum_20190213 = klo1.groupby('WorkstationID').sum().sort_values(by='ArtNum',ascending = False).reset_index()
sum_of_ArtNum_20190328 = klo2.groupby('WorkstationID').sum().sort_values(by='ArtNum',ascending = False).reset_index()
```

#เเยกวันเดือน ปี เวลา

```python
concot_TN = pd.read_csv('/content/drive/MyDrive/สอบกลางภาค/data use/3 time ter มีrating.csv')
```

```python
cons_id1 = concot_TN['BeginDateTime'].tolist()
cons_id2 = concot_TN['EndDateTime'].tolist()
cons_no = []
cons_end = []
date_S=[]
date_E=[]
for i in range(len(cons_id1)):
  re = cons_id1[i].split('T')[-1]
  cons_no.append(re)
for i in range(len(cons_id2)):
  re = cons_id2[i].split('T')[-1]
  cons_end.append(re)
for i in range(len(cons_id1)):
  re = cons_id2[i].split('T')[0]
  date_S.append(re)
for i in range(len(cons_id2)):
  re = cons_id2[i].split('T')[0]
  date_E.append(re)
concot_TN['date_S'] = date_S
concot_TN['date_E'] = date_E
concot_TN['time_S'] = cons_no
concot_TN['time_E'] = cons_end
concot_TN
```
